```python
#Importing Libraries
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

## Function
## $f(x) = x^2 + x + 1 $

```python
def f(x):
    return x**2 + x + 1
```

```python
#Generating Data
x = np.linspace(start=-4, stop=4, num=500)
```

```python
#Plotting the Function
plt.figure(figsize=(10,8))
plt.plot(x, f(x), color='blue', linewidth=4)
plt.xlabel('x')
plt.ylabel('f(x) = x^2 + x + 1')
plt.title('f(x) v/s x')
plt.show()
```

## Derivative of f(x)
## $ df/dx= 2x + 1$

```python
#Derivative of f(x)
def df(x):
    return 2*x + 1
```

```python
#Plotting Cost Function and it's Derivative
plt.figure(figsize=(15, 5))

#Cost Function
plt.subplot(1, 2, 1)
plt.plot(x, f(x), color='blue', linewidth=4)
plt.xlabel('x')
plt.ylabel('f(x) = x^2 + x + 1')
plt.title('f(x) v/s x')

#Derivative of Cost Function
plt.subplot(1, 2, 2)
plt.plot(x, df(x), color='red', linewidth=4)
plt.grid()
plt.xlabel('x')
plt.ylabel('df(x) = x*2 + 1')
plt.title('df(x) v/s x')
plt.show()
```

## Gradient Descent

```python
def gradient_descent(derivative_func, x_init=0, learning_rate=1e-2, tolerance=1e-4, num_iterations=500):
    current_x = x_init
    x_list = [current_x]
    slope_list = [derivative_func(current_x)]

    for i in range(num_iterations):
        previous_x = current_x
        gradient = derivative_func(previous_x)
        current_x = previous_x - learning_rate * gradient
        gradient = derivative_func(current_x)
        x_list.append(current_x)
        slope_list.append(gradient)
        step = abs(current_x - previous_x)
        if step < tolerance:
            break
    x_list = np.array(x_list)
    slope_list = np.array(slope_list)
    return current_x, x_list, slope_list, i
```

```python
num_iterations = 500
tolerance = 1e-4
learning_rate = 1e-2
x_init = 4

optimum_x, x_list, slope_list, actual_iterations = gradient_descent(derivative_func=df, x_init=x_init, 
                                                                    learning_rate=learning_rate, 
                                                                    tolerance=tolerance, 
                                                                    num_iterations=num_iterations)

print(f"Cost Function is minimum at x = {optimum_x}.")
print(f"Minimum Value of Cost Function is {f(optimum_x)}.")
print(f"Value of Slope of Cost Function at Cost Function minimum is {df(optimum_x)}")
print(f"Gradient Descent ran for {actual_iterations} iterations.")
```

```python
#Plot of Gradient Descent 
plt.figure(figsize=(15, 5))

#Cost Function
plt.subplot(1, 2, 1)
plt.plot(x, f(x), color='blue', linewidth=4, alpha=0.4)
plt.scatter(x_list, f(x_list), color='red', s=50, alpha=0.3)
plt.scatter(optimum_x, f(optimum_x), color='black', s=60)
plt.xlabel('x')
plt.ylabel('f(x) = x^2 + x + 1')
plt.title('f(x) v/s x')

#Derivative of Cost Function
plt.subplot(1, 2, 2)
plt.plot(x, df(x), color='blue', linewidth=4, alpha=0.4)
plt.scatter(x_list, slope_list, color='red', s=50, alpha=0.3)
plt.scatter(optimum_x, df(optimum_x), color='black', s=60)
plt.grid()
plt.xlabel('x')
plt.ylabel('df(x) = x*2 + 1')
plt.title('df(x) v/s x')
plt.show()
```
