```python
#Importing Libraries
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

## Function
## $f(x) = x^4 - 4x^2 + 5 $

```python
def f(x):
    return x**4 -4*x**2 + 5
```

## Derivative of f(x)
## $ df/dx= 4x^3 -8x$

```python
def df(x):
    return 4*x**3 - 8*x
```

```python
#Generating Data
x = np.linspace(start=-4, stop=4, num=500)
```

```python
#Plotting Cost Function and it's Derivative
plt.figure(figsize=(15, 5))

#Cost Function
plt.subplot(1, 2, 1)
plt.plot(x, f(x), color='blue', linewidth=4)
plt.xlabel('x')
plt.ylabel('f(x) = x^4 - 4*x^2 + 5')
plt.title('f(x) v/s x')

#Derivative of Cost Function
plt.subplot(1, 2, 2)
plt.plot(x, df(x), color='red', linewidth=4)
plt.grid()
plt.xlabel('x')
plt.ylabel('df(x) = 4*x^3 - 8*x')
plt.title('df(x) v/s x')
plt.show()
```

## Gradient Descent

```python
def gradient_descent(derivative_func, x_init=0, learning_rate=1e-2, tolerance=1e-4, num_iterations=500):
    current_x = x_init
    x_list = [current_x]
    slope_list = [derivative_func(current_x)]

    for i in range(num_iterations):
        previous_x = current_x
        gradient = derivative_func(previous_x)
        current_x = previous_x - learning_rate * gradient
        gradient = derivative_func(current_x)
        x_list.append(current_x)
        slope_list.append(gradient)
        step = abs(current_x - previous_x)
        if step < tolerance:
            break
    x_list = np.array(x_list)
    slope_list = np.array(slope_list)
    return current_x, x_list, slope_list, i
```

```python
num_iterations = 500
tolerance = 1e-4
learning_rate = 1e-2
```

## x_init = -3

```python
x_init_1 = -3

optimum_x_1, x_list_1, slope_list_1, actual_iterations_1 = gradient_descent(derivative_func=df, x_init=x_init_1, 
                                                                    learning_rate=learning_rate, 
                                                                    tolerance=tolerance, 
                                                                    num_iterations=num_iterations)

print(f"Initial x value is {x_init_1}.")
print(f"Cost Function is minimum at x = {optimum_x_1}.")
print(f"Minimum Value of Cost Function is {f(optimum_x_1)}.")
print(f"Value of Slope of Cost Function at Cost Function minimum is {df(optimum_x_1)}")
print(f"Gradient Descent ran for {actual_iterations_1} iterations.")
```

## x_init = 3

```python
x_init_2 = 3

optimum_x_2, x_list_2, slope_list_2, actual_iterations_2 = gradient_descent(derivative_func=df, x_init=x_init_2, 
                                                                    learning_rate=learning_rate, 
                                                                    tolerance=tolerance, 
                                                                    num_iterations=num_iterations)
print(f"Initial x value is {x_init_2}.")
print(f"Cost Function is minimum at x = {optimum_x_2}.")
print(f"Minimum Value of Cost Function is {f(optimum_x_2)}.")
print(f"Value of Slope of Cost Function at Cost Function minimum is {df(optimum_x_2)}")
print(f"Gradient Descent ran for {actual_iterations_2} iterations.")
```

## x_init = 0.1

```python
x_init_3 = 0.1

optimum_x_3, x_list_3, slope_list_3, actual_iterations_3 = gradient_descent(derivative_func=df, x_init=x_init_3, 
                                                                    learning_rate=learning_rate, 
                                                                    tolerance=tolerance, 
                                        
                                                                            num_iterations=num_iterations)
print(f"Initial x value is {x_init_3}.")
print(f"Cost Function is minimum at x = {optimum_x_3}.")
print(f"Minimum Value of Cost Function is {f(optimum_x_3)}.")
print(f"Value of Slope of Cost Function at Cost Function minimum is {df(optimum_x_3)}")
print(f"Gradient Descent ran for {actual_iterations_3} iterations.")
```

```python
#Plot of Gradient Descent with different initial values
plt.figure(figsize=(15, 5))

#Cost Function
plt.subplot(1, 2, 1)
plt.plot(x, f(x), color='blue', linewidth=4, alpha=0.4)

plt.scatter(x_list_1, f(x_list_1), color='red', s=50, alpha=0.6)
plt.scatter(optimum_x_1, f(optimum_x_1), color='black', s=60)

plt.scatter(x_list_2, f(x_list_2), color='orange', s=50, alpha=0.6)
plt.scatter(optimum_x_2, f(optimum_x_2), color='black', s=60)

plt.scatter(x_list_3, f(x_list_3), color='green', s=50, alpha=0.6)
plt.scatter(optimum_x_3, f(optimum_x_3), color='black', s=60)

plt.xlim(-3.5, 3.5)
plt.ylim(-2, 75)
plt.xlabel('x')
plt.ylabel('f(x) = x^4 - 4*x^2 + 5')
plt.title('f(x) v/s x')

#Derivative of Cost Function
plt.subplot(1, 2, 2)
plt.plot(x, df(x), color='blue', linewidth=4, alpha=0.4)

plt.scatter(x_list_1, slope_list_1, color='red', s=50, alpha=0.6)
plt.scatter(optimum_x_1, df(optimum_x_1), color='black', s=60)

plt.scatter(x_list_2, slope_list_2, color='orange', s=50, alpha=0.6)
plt.scatter(optimum_x_2, df(optimum_x_2), color='black', s=60)

plt.scatter(x_list_3, slope_list_3, color='green', s=50, alpha=0.6)
plt.scatter(optimum_x_3, df(optimum_x_3), color='black', s=60)

plt.grid()
plt.xlim(-3.5, 3.5)
plt.ylim(-100, 100)
plt.xlabel('x')
plt.ylabel('df(x) = 4*x^3 - 8*x')
plt.title('df(x) v/s x')
plt.show()
```
